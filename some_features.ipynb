```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

path= '/content/drive/My Drive/NBA/player_condition.csv'

df= pd.read_csv(path)
```

```python
df['GAME_DATE_EST']=pd.to_datetime(df['GAME_DATE_EST'])
df= df.sort_values(by=['GAME_DATE_EST', 'PLAYER_NAME'])
```

```python
#DAYS SINCE LAST GAME FEATURE
df['DAYS_SINCE_LG']= df.groupby(['PLAYER_NAME', 'SEASON'])['GAME_DATE_EST'].diff().dt.days.fillna(0)

df[(df['PLAYER_NAME']=='LeBron James') & (df['SEASON']==17)]
```

```python
#CUMULATIVE WORKLOAD: #Games played in the season since a certain game
df['CUMULATIVE_WL']= df.groupby(['PLAYER_NAME', 'SEASON']).cumcount()
```

```python
##IS STARTER VAR
df['START_RATIO']= df['GS']/ df['GP']

df['IS_STARTER'] = (df['START_RATIO'] > 0.5).astype(int)

df.drop(columns=['START_RATIO'], inplace=True)

display(df[df['PLAYER_NAME']=='LeBron James'].head())
```

```python
#Back to Back variable

df['BTB']= (df['DAYS_SINCE_LG']==1).astype(int)
```

```python
# ROAD_GAMES_STREAK: numero di partite consecutive fuori casa
import pandas as pd

def calculate_road_streak(home_game_series):
    streaks = []
    current_streak = 0
    for is_home in home_game_series:
       if is_home == 0:
           current_streak += 1
       else:
           current_streak = 0
       streaks.append(current_streak)
    return pd.Series(streaks, index=home_game_series.index)


df['ROAD_GAMES_STREAK'] = df.groupby(['PLAYER_NAME', 'SEASON'])['HOME_GAME'].transform(calculate_road_streak)
```
